## Unified API 
- Unified API to process both batch and streaming data.
- Single abstraction (**PCollection**) and operators (**ParDo, GroupByKey, Window**, etc.) for both bounded (**batch**) and unbounded (**streaming**) data.  

- **Ex:** Spark/Flink → Separate API for both Batch and Streaming data.  
  - **Spark** → `DataFrame` for Batch, `readStream` for Streaming.  
  - **Flink** → `DataSet` for Batch, `DataStream` for Streaming.

Before beam how destributed framework evolved over the time.
in 2004 google published mapReduce paper which fundamentally changed the way we process destributed data 
basically in map reduce the whole process is getting done in two steps.
a map and reduce phase.
Now google took this MapReduce and continued to innovate more systems and papers over the time but they are not the open source products and were not initially released pubically, they were only being used at google internally within their systems.
and meanwhile at same time, open source communities like Apache took that MapReduce paper and created hadoop which was like a revolution in the Big data processing frameworks.
Hadoop used MapReduce as base and started doing distributed processing as key value pairs.
now even though it managed to provide Scalability and simplicity in distributed processing world using low cost machines.
But sooner we realized that at core mapReduce is just a single select and groudby .
MapReduce worked well for the use cases which can be broken into map and reduce phase but for many real-world use cases, a hain of map and reduce stages is required.
A chain of MapReduce jobs mean a lot of disk reads and writes, apparently increasing the overall processing time.
that is why there was a need of high-level API that can create pipeline of operations with less hdfs read and writes.
that what from where Flume arose.
Flume basically was the Orchestration and optimization of MapReduce.
Rather that a single operation, flume creates a graph of multiple operations which eventually get converted into an optimize serie of mapReduces without multiple hdfs writes.
Actualy, later on it was flune which emerged as the base programming model for Spark and Flink as well.
Then we have Spark, a lightening fast processing engine which they promise to be 100 times faster than hadoop.
it supported both batch and stream processing.
at core it does lazy evaluation of the program creates a DAG of the code and then excecute it.
Spark streaming is also fast but it does not have any mechanism to handle late elements or we can say out od elements in window.
So, Spark was not prepared to handle our od order elements which was later on addressed by a new Apache product i.e. is Flink 
Flink was more or less same like Spark but with more speed, performance and automated tuning.
it was actually the first engine which started considering a data strean as first class citizen and that is why it became a true stream processing engine.
but the major thing ws its ability to handle out od order elements in a stream.
it had lot of mechanisems llike watermarking, states, triggers ets to do so.

Now as you can see there are 2 seperate worled flourisheing in thier own zone.
these zones were isolated from each other and the compoenents in them cannot integrate with each other.
Means you cannot just create a spark scala code and run it using google cloud dataflow or vice versa, code created for google MIllwheel cannot be run on Apache engines.
Now what beam is doib, beam is bringing these 2 worlds together.
it is based on dataflow programing model and generalising integrated with a broder big data ecosystem.
after this integration , you can now build language agnostic pipeline using Apache beam programing model and run it anywhere either Open source runners or Google cloud dataflow.
Basically it means if you write a code in Beam then it run on any of the Google's or Apache's engines.
Moreover, Beam has also seperated the tight integration of language SDKs and its runners.
With this you can choose any of the programming language that Beam supports, bulid pipelines using that language and run it on any engine.
So create a beam pogram in any of your favorite language and run that pogram in any of your favorite engine.
